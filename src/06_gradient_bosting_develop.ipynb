this time we will use the gradient boosting from thew library xgboost to predict the diabetes from medical data

```python
import pandas as pd

# load the csv data sets

# Con Index = False, no guardamos el indice en el archivo resultante
X_train = pd.read_csv("../data/processed/X_train.csv")
X_test = pd.read_csv("../data/processed/X_test.csv")

y_train = pd.read_csv("../data/processed/y_train.csv")
y_test = pd.read_csv("../data/processed/y_test.csv")
```

```python
import torch

# Verificar si CUDA está disponible
if torch.cuda.is_available():
    # Obtener el número de dispositivos CUDA disponibles
    num_gpu = torch.cuda.device_count()

    print(f"Número de GPUs con soporte CUDA encontradas: {num_gpu}")

    # Listar información detallada sobre cada GPU
    for i in range(num_gpu):
        gpu_info = torch.cuda.get_device_properties(i)
        print(f"\nGPU {i + 1}:\n")
        print(f"Nombre: {gpu_info.name}")
        print(f"Capacidad de cómputo: {gpu_info.major}.{gpu_info.minor}")
        print(f"Memoria global: {gpu_info.total_memory / (1024 ** 3):.2f} GB")
else:
    print("No se encontraron dispositivos CUDA en tu sistema.")
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(random_state = 42, device= 'cuda')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# Obténgo la lista de parámetros válidos y sus valores actuales
params = model.get_params()

# Guardo esos valores en una lista
lista_params = list(params.keys())
lista_params
```

```python
import xdcodes.scrapgpt as xd

gptdriver = xd.open_driver()
```

```python
xd.new_chat(gptdriver)
```

```python

explicacion_parametros = []


for prm in lista_params:

    prompt = f'que es {prm} en xgboost resumelo en una frase'

    respuesta = xd.scrap_gpt(gptdriver, prompt)

    explicacion_parametros.append(respuesta)
    print(respuesta)
```

```python
# Especifica la ruta del archivo y el modo de apertura (en este caso, escritura 'w')
ruta_archivo = "../docs/explicacion_parametros.txt"

with open(ruta_archivo, "w") as archivo:
    # Escribe cada elemento de la lista en una nueva línea
    for elemento in explicacion_parametros:
        archivo.write(f"- {elemento}\n")

print(f"Contenido guardado en {ruta_archivo}")
```

hyperparametrization with xgboosting and scikitlearn

```python
# A parameter grid for XGBoost
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5],
        }
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV


grid = GridSearchCV(model, param_grid=params, scoring='accuracy', n_jobs=-2, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
# Print the best set of hyperparameters and the corresponding score
print("Best set of hyperparameters: ", grid.best_params_)
print("Best score: ", grid.best_score_)
```

```python
model = XGBClassifier( **grid.best_params_ , random_state = 42, device= 'cuda')
model.fit(X_train, y_train)
```

```python
import pickle

#Cambiar el objeto por el que queramos guardar
objeto = model

ruta_guardado = '../models/boosting_classifier.pkl'

with open(ruta_guardado, "wb") as archivo:
    pickle.dump(objeto, archivo)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
